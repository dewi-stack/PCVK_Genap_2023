8. Lengkapi tulisan nama file pada file citra dari soal no.8. gunakan font, ukuran font, dan warna font yang sesuai keinginan anda. 

```python
img = cv.imread('/content/drive/MyDrive/PCVK/img/kitten.jpg')
fig = plt.figure(figsize=(5,5))

ax = fig.add_subplot(111)
cv.rectangle(img, pt1=(150,550), pt2=(700,100), color=(255, 0, 30), thickness = 40)
cv.circle(img, center=(412, 350), radius=260, color=(0, 255, 0), thickness = 30)
ax.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))

font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(0,0,255),thickness=10,lineType=cv.LINE_AA)
plt.imshow(img)
```
